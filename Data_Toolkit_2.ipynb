```python
# Create a time-series dataset in a Pandas DataFrame with columns: 'Date', 'Temperature', 'Humidity' and
# Perform the following tasks using Matplotlib:

# a) Plot the 'Temperature' and 'Humidity' on the same plot with different y-axes (left y-axis for 'Temperature' and
# right y-axis for 'Humidity').

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Generate a date range
date_range = pd.date_range(start='2023-01-01', periods=100, freq='D')

# Create a DataFrame with random Temperature and Humidity values
np.random.seed(0)  # For reproducibility
data = {
    'Date': date_range,
    'Temperature': np.random.normal(loc=20, scale=5, size=len(date_range)),  # Random temperatures around 20C
    'Humidity': np.random.normal(loc=60, scale=10, size=len(date_range))     # Random humidity around 60%
}

df = pd.DataFrame(data)

# Display the DataFrame
print(df.head())

# Create a figure and a set of subplots
fig, ax1 = plt.subplots()

# Plot Temperature on the left y-axis
ax1.plot(df['Date'], df['Temperature'], 'r-', label='Temperature')
ax1.set_xlabel('Date')
ax1.set_ylabel('Temperature (C)', color='r')
ax1.tick_params(axis='y', labelcolor='r')


# Rotate x-axis labels to avoid overlapping
plt.xticks(rotation=45)

# Create a second y-axis sharing the same x-axis
ax2 = ax1.twinx()
ax2.plot(df['Date'], df['Humidity'], 'b-', label='Humidity')
ax2.set_ylabel('Humidity (%)', color='b')
ax2.tick_params(axis='y', labelcolor='b')

# Adding title and legends
plt.title('Temperature and Humidity Over Time')
fig.tight_layout()  # Adjust layout to prevent clipping
plt.show()
```

```python
# Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following
# tasks using Matplotlib:
# a) Plot a histogram of the data with 30 bins.

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram of 1000 Samples from a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# b) Overlay a line plot representing the normal distribution's probability density function (PDF).

import numpy as np
import matplotlib.pyplot as plt

# Generate 1000 samples from a normal distribution
data = np.random.normal(loc=0, scale=1, size=1000)

# Plot a histogram of the data with 30 bins
plt.hist(data, bins=30, density=True, edgecolor='black', alpha=0.7, label='Histogram')

# Overlay a line plot representing the PDF of the normal distribution
mu, sigma = 0, 1  # mean and standard deviation
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
pdf = (1/(sigma * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mu) / sigma)**2)
plt.plot(x, pdf, color='red', linewidth=2, label='PDF')

plt.title('Histogram and PDF of 1000 Samples from a Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density/Probability')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#  Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
# origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate random data
x = np.random.randn(100)
y = np.random.randn(100)

# Determine quadrant for each point
quadrant = np.zeros_like(x, dtype=int)
quadrant[(x > 0) & (y > 0)] = 1
quadrant[(x < 0) & (y > 0)] = 2
quadrant[(x < 0) & (y < 0)] = 3
quadrant[(x > 0) & (y < 0)] = 4

# Create a scatter plot with Seaborn
sns.scatterplot(x=x ,y=y ,hue=quadrant, palette='Set1', legend='full')

# Add legend, labels, and title
plt.legend(title='Quadrant')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

plt.show()
```

```python
# With Bo+eh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine
# Wave Function'8.

import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(0, 2 * np.pi, 1000)

# Compute the sine of x
y = np.sin(x)

# Create the plot
plt.plot(x, y, label='Sine wave')

# Add grid lines
plt.grid(True)

# Label the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Set the title
plt.title('Sine Wave Function')

# Display the legend
plt.legend()

# Show the plot
plt.show()


```

```python
# Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
# values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
# Bar Chart'8

from bokeh.io import show, output_notebook
from bokeh.plotting import figure
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
import numpy as np
import pandas as pd

# Display plots inline in Jupyter Notebook
output_notebook()

# Generate random categorical data
categories = ['A', 'B', 'C', 'D', 'E']
values = np.random.randint(10, 100, size=len(categories))

# Create a DataFrame
data = pd.DataFrame({'categories': categories, 'values': values})

# Create a ColumnDataSource
source = ColumnDataSource(data)

# Define a color mapper
colors = ["#718dbf", "#e84d60", "#ddb7b1", "#a6bddb", "#2b83ba"]
color_mapper = factor_cmap('categories', palette=colors, factors=categories)

# Create the figure
p = figure(x_range=categories, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars to the plot
p.vbar(x='categories', top='values', width=0.9, source=source,
       line_color='white', fill_color=color_mapper)

# Add hover tool
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Label the axes
p.xaxis.axis_label = 'Categories'
p.yaxis.axis_label = 'Values'

# Show the plot
show(p)
```

```python
# Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
# 'Simple Line Plot'8.


import plotly.graph_objects as go
import numpy as np

# Generate random dataset
x = np.linspace(0, 10, 100)
y = np.random.randn(100)

# Create the figure
fig = go.Figure()

# Add a line plot
fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Label the axes
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()

```

```python
# Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
# the title as 'Interactive Pie Chart'.

import plotly.graph_objects as go
import numpy as np

# Generate random data
np.random.seed(0)  # For reproducibility
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(categories))

# Create the pie chart
fig = go.Figure(data=[go.Pie(labels=categories, values=values, hole=.3, 
                             textinfo='label+percent', insidetextorientation='radial')])

# Update layout
fig.update_layout(title_text='Interactive Pie Chart')

# Show the plot
fig.show()
```
